# Fit gamma distribution

```python
x = np.linspace(0,300,1000)

# Gamma
shape, loc, scale = stats.gamma.fit(data, floc=0)

print(shape, loc, scale)
y = stats.gamma.pdf(x, shape, loc, scale)
plt.title('Fitted Gamma')
plt.plot(x, y)
plt.show()
```

# Fit inverse gamma distribution

```python
x = np.linspace(0,300,1000)

params = stats.invgamma.fit(data, floc=0)

y = stats.invgamma.pdf(x, *params)
plt.title('Fitted Inverse Gamma')
plt.plot(x, y)
plt.show()
```

# Fit lognormal distribution

```python
x = np.linspace(0,300,1000)

params = stats.lognorm.fit(data, floc=0)

y = stats.lognorm.pdf(x, *params)
plt.title('Fitted Lognormal')
plt.plot(x, y)
plt.show()
```

# Create and fit lognormal distribution

```python
sample = stats.lognorm.rvs(loc=0, s=1, size=1000, random_state=0)

# plot sample
x_axis = np.linspace(0,10,1000)

plt.figure(figsize=(12,8))
plt.plot(x_axis, sample,'b.')


# fit lognormal
params = stats.lognorm.fit(sample, floc=0)
y_axis = stats.lognorm.pdf(x_axis, *params)

# plot
plt.title('Fitted Lognormal')
plt.plot(x_axis, y_axis)
plt.show()
```
